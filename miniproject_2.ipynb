```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_India","Active cases")
```

```python
len(df)
```

```python
df[2]
```

```python
state = df[2].iloc[0:36,1]
cases = list(map(int,df[2].iloc[0:36,2]))
recovered = list(map(int,df[2].iloc[0:36,4]))
```

```python
plt.figure(figsize=(80,30))
plt.bar(state,cases,label="India",width=0.4)
plt.bar(state+0.4,recovered,label="Infected",width=0.4)
plt.legend()
plt.show()
```
